```python
import pandas as pd

def generate_group_label_synonym_dict(pd_file):
    """
    生成标签分组 -> 标签 -> 同义词 的映射字典，去除索引信息，保留标签分组和标签的原始值。

    参数:
    pd_file (str): 输入CSV文件的路径。

    返回:
    dict: 返回包含标签分组、标签和同义词的嵌套字典。
    """
    # 读取CSV文件
    dict_data = pd.read_csv(pd_file)

    # 创建"标签分组" -> "标签" -> "同义词"的嵌套映射字典
    group_label_synonym_dict = {}

    # 按照"标签分组"进行分组
    for group in dict_data['标签分组'].unique():
        # 获取该标签分组下的所有"标签"
        group_data = dict_data[dict_data['标签分组'] == group]

        # 为该标签分组下创建嵌套字典
        group_label_synonym_dict[group] = {'labels': {}}

        # 为该标签分组下的标签创建字典
        for label in group_data['标签'].unique():
            group_label_synonym_dict[group]['labels'][label] = {'synonyms': {}}

            # 为该标签下的同义词分配
            for synonym in group_data[group_data['标签'] == label]['同义词'].unique():
                group_label_synonym_dict[group]['labels'][label]['synonyms'][synonym] = synonym

    return group_label_synonym_dict
```

```python
pd_file = '/data1/code/dengxinzhe/sources/Flux设计要素词汇表 - 触发词同义词表-20250208.csv'
group_label_synonym_dict = generate_group_label_synonym_dict(pd_file)

print(group_label_synonym_dict)
```
